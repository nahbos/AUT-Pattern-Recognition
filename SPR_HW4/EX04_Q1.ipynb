```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0.4, 1.5, 1.6, 2.2, 2.4, 2.8, 3.4, 3.8, 4.0, 4.2],
             [4.5, 0.8, 1.4, 4.2, 1.0, 0.2, 1.8, 3.6, 2.3, 2.9]])
centers = np.array([[2.2, 2.0, 1.2],
                    [2.8, 2.0, 3.0]])
```

```python
x_labels = []
c_labels = []
for ele in x.T:
    x_labels.append('(' + str(ele[0]) + ', ' + str(ele[1]) + ')')
    
for ele in centers.T:
    c_labels.append('(' + str(ele[0]) + ', ' + str(ele[1]) + ')')
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(x[0], x[1], c='gray', s=100)

for i, txt in enumerate(x_labels):
    plt.annotate(txt, (x[0][i], x[1][i]), xytext=(x[0][i] + 0.1, x[1][i] + 0.1))

plt.scatter(centers[0][0], centers[1][0], c='red', marker='*', s=150)
plt.scatter(centers[0][1], centers[1][1], c='blue', marker='*', s=150)
plt.scatter(centers[0][2], centers[1][2], c='green', marker='*', s=150)

for i, txt in enumerate(c_labels):
    plt.annotate(txt, (centers[0][i], centers[1][i]), xytext=(centers[0][i] + 0.1, centers[1][i] + 0.1), c=['red', 'blue', 'green'][i])

plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
for j in range(3):
    print('--------------  ', ['red', 'blue', 'green'][j], '  --------------')
    for i in range(x.T.shape[0]):
        print(x.T[i], ':')
        print(round(np.linalg.norm(np.subtract(x.T , centers.T[j])[i]), 4), '\n')
```

```python
iter1_centers = np.array([[3.55, 2.34, 0.4],
                          [3.25, 1.04, 4.5]])

iter1_c_labels = []
for ele in iter1_centers.T:
    iter1_c_labels.append('(' + str(ele[0]) + ', ' + str(ele[1]) + ')')
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(x[0], x[1], c='gray', s=100)

for i, txt in enumerate(x_labels):
    plt.annotate(txt, (x[0][i], x[1][i]), xytext=(x[0][i] + 0.1, x[1][i] + 0.1))

plt.scatter(iter1_centers[0][0], iter1_centers[1][0], c='red', marker='*', s=150)
plt.scatter(iter1_centers[0][1], iter1_centers[1][1], c='blue', marker='*', s=150)
plt.scatter(iter1_centers[0][2], iter1_centers[1][2], c='green', marker='*', s=150)

for i, txt in enumerate(iter1_c_labels):
    plt.annotate(txt, (iter1_centers[0][i], iter1_centers[1][i]), 
                 xytext=(iter1_centers[0][i] - 0.35, iter1_centers[1][i] + 0.15), c=['red', 'blue', 'green'][i])

plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```
