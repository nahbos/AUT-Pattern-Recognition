```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

#### Reading data

```python
iris = datasets.load_iris()
dataset = pd.DataFrame(data= np.c_[iris.data, iris.target], columns= iris.feature_names + ['target'])
dataset.head()
```

```python
dataset.info()
```

#### Ok. Letmme do a group by on the dataset for having 3 classes.

```python
figs = [(1, 2), (1, 3), (2, 4)]
T = [(1.2, -0.3), (-1.8, 0.6), (1.4, 0.5), (-0.5, -1)]

for j in range(len(T)):
    fig, ax = plt.subplots(3, 1, figsize=(10, 12))
    
    for i in range(3):
        
        dataset['trans'] = dataset.iloc[:, figs[i][0] - 1] * T[j][0] + dataset.iloc[:, figs[i][1] - 1] * T[j][1]
        classes = dataset.groupby(by=dataset.loc[:, 'target'])

        for grp, pdf in classes:
            ax[i].scatter(pdf.iloc[:, -1], [5] * len(pdf.iloc[:, -1]), alpha=0.5, label='class {}'.format((int(grp) + 1)))
            ax[i].legend()
            ax[i].grid()
            ax[i].set_title('Distribution of Feature {} and {}'.format(figs[i][0], figs[i][1]))
```
